```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = "/content/drive/MyDrive/household_power_consumption.xlsx"
df = pd.read_excel(file_path)
```

```python
sub_df = df[(df['Date'] >= '2007-02-01') & (df['Date'] <= '2007-02-02')]
```

```python
sub_df['Datetime'] = pd.to_datetime(sub_df['Date'].astype(str) + " " + sub_df['Time'].astype(str))
```

```python

plt.figure()
plt.plot(sub_df['Datetime'], sub_df['Sub_metering_1'], 'k-', label='Sub_metering_1')
plt.plot(sub_df['Datetime'], sub_df['Sub_metering_2'], 'r-', label='Sub_metering_2')
plt.plot(sub_df['Datetime'], sub_df['Sub_metering_3'], 'b-', label='Sub_metering_3')
plt.ylabel('Global Active Power (kilowatts)')
plt.xlabel('')
plt.legend(loc='upper right')
plt.show()
```
