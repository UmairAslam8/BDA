```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = "/content/drive/MyDrive/household_power_consumption.xlsx"
df = pd.read_excel(file_path)
```

```python
sub_df = df[(df['Date'] >= '2007-02-01') & (df['Date'] <= '2007-02-02')]
sub_df['Datetime'] = pd.to_datetime(sub_df['Date'].astype(str) + " " + sub_df['Time'].astype(str))
```

```python

plt.figure(figsize=(12, 6))
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.4, hspace=0.4)

plt.subplot(2, 2, 1)
plt.plot(sub_df['Datetime'], sub_df['Global_active_power'], 'k-')
plt.ylabel('Global Active Power (kilowatts)')

plt.subplot(2, 2, 2)
plt.plot(sub_df['Datetime'], sub_df['Voltage'], 'k-')
plt.ylabel('Voltage (volt)')

plt.subplot(2, 2, 3)
plt.plot(sub_df['Datetime'], sub_df['Sub_metering_1'], 'k-')
plt.plot(sub_df['Datetime'], sub_df['Sub_metering_2'], 'r-')
plt.plot(sub_df['Datetime'], sub_df['Sub_metering_3'], 'b-')
plt.ylabel('Global Active Power (kilowatts)')
plt.legend(['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3'], loc='upper right', frameon=False)

plt.subplot(2, 2, 4)
plt.plot(sub_df['Datetime'], sub_df['Global_reactive_power'], 'k-')
plt.ylabel('Global Rective Power (kilowatts)')
plt.savefig('plot4.png', dpi=100, bbox_inches='tight')
```
