```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = "/content/drive/MyDrive/household_power_consumption.xlsx"
df = pd.read_excel(file_path)
```

```python
sub_df = df[(df['Date'] >= '2007-02-01') & (df['Date'] <= '2007-02-02')]
```

```python

sub_df['Datetime'] = pd.to_datetime(sub_df['Date'].astype(str) + " " + sub_df['Time'])
sub_df['DayOfWeek'] = sub_df['Datetime'].dt.strftime('%A')

# Plotting
sub_df.plot(x='DayOfWeek', y='Global_active_power', kind='line', color='blue')
plt.savefig('plot2.png', dpi=100, bbox_inches='tight')
plt.show()
```
