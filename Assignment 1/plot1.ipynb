```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = "/content/drive/MyDrive/household_power_consumption.xlsx"
df = pd.read_excel(file_path)
```

```python
sub_df = df[(df['Date'] >= '2007-02-01') & (df['Date'] <= '2007-02-02')]
```

```python
sub_df['Global_active_power'] = pd.to_numeric(sub_df['Global_active_power'], errors='coerce')

# Remove rows with missing values in 'Global Active Power' column
sub_df = sub_df.dropna(subset=['Global_active_power'])

plt.hist(sub_df['Global_active_power'],bins=18, color='red')

# Add labels and title
plt.xlabel('Global Active Power (Kilowatts)')
plt.ylabel('Frequency')
plt.title('Histogram of Global Active Power')


# Save the plot as a PNG file with specified dimensions
plt.savefig('plot1.png', dpi=100, bbox_inches='tight')

# Show the plot
plt.show()
```
